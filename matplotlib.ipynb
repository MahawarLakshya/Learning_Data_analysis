Axeslabels ,legenda and grid

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title("Weather")
plt.plot(days,max_t,label='maximum temp')
plt.plot(days,min_t,label='minimum_temp')
plt.plot(days,avg_t,label='average_temp')
plt.legend()
```

```python
#to change the location of legend use loc property of legend
#if location is not specified it will identify the blank area and make it place there
plt.xlabel('Days')
plt.ylabel('TEmperature')
plt.title("Weather")
plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label='minp')
plt.plot(days,avg_t,label='avg')
plt.legend(loc='upper right')
```

```python
#shadow and font size property of legend
plt.xlabel('Days')
plt.ylabel('TEmperature')
plt.title("Weather")
plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label='min')
plt.plot(days,avg_t,label='avg')
plt.legend(loc='upper right',shadow=True,fontsize='small')
```

```python
#use grid for better visualization
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label='min')
plt.plot(days,avg_t,label='avg')
plt.legend(loc='upper right',shadow=True)
plt.grid()
```

BAR CHART image.png

```python

import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
xpos=np.arange(len(company))
xpos
#to replace the array value as numbers to other value use xticks
plt.xticks(xpos,company)
plt.xlabel('US Companies stocks')
plt.ylabel('Revenue(bln)')
plt.bar(xpos,revenue)#it only accepts string values
```

```python
#to add two different measures in bar char
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,35,13]
```

```python

xpos=np.arange(len(company))
xpos
#to replace the array value as numbers to other value use xticks
plt.xticks(xpos,company)
plt.xlabel('US Companies stocks')
plt.ylabel('Revenue(bln)')
plt.bar(xpos,revenue,label='Revenue')#it only accepts string values
plt.bar(xpos,profit,label='Profit')
plt.legend()
```

```python
#above code generate the plot in same bar,as we have same xpos,if we want to generate it side by side we do it by substracting xpos woth some number
xpos=np.arange(len(company))
xpos
#to replace the array value as numbers to other value use xticks
plt.xticks(xpos,company)
plt.xlabel('US Companies stocks')
plt.ylabel('Revenue(bln)')
plt.bar(xpos+0.2,revenue,label='Revenue')#it only accepts string values
plt.bar(xpos-0.2,profit,label='Profit')
plt.legend()
```

```python
#we can also control the width of the bar by using width property of bar
xpos=np.arange(len(company))
xpos
#to replace the array value as numbers to other value use xticks
plt.xticks(xpos,company)
plt.xlabel('US Companies stocks')
plt.ylabel('Revenue(bln)')
plt.bar(xpos-.2,revenue,width=0.4,label='Revenue')#it only accepts string values
plt.bar(xpos+.2,profit,width=0.4,label='Profit')
plt.legend()
```

```python
#to make the bar chart in horizontal fashion we use barh instead of bar
xpos=np.arange(len(company))
xpos
#to replace the array value as numbers to other value use xticks
plt.xticks(xpos,company)
plt.xlabel('US Companies stocks')
#to make the bar chart in horizontal fashion we use barh instead of bar
xpos=np.arange(len(company))
xpos
#to replace the array value as numbers to other value use xticks
plt.xticks(xpos,company)
plt.xlabel('US Companies stocks')
plt.ylabel('Revenue(bln)')
plt.barh(xpos-.2, revenue, height=0.4, label='Revenue')#it only accepts string values
plt.barh(xpos+.2, profit,height=0.4, label='Profit')
plt.legend()
```

Histogram are the bar chart where y axis cconatins the frequency of variable
image.png

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar=[113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar)
```

```python
#by default it will give you 10 bins in histogram we have concept of bucket where we can set the range if we want to add 3 bins in itwe put
plt.hist(blood_sugar,bins=3)
```

```python
#to unjoin the bars we use rwidth parameter which is relative bin comapred to the bin
plt.hist(blood_sugar,bins=3,rwidth=0.95)#.95 of entire range
```

```python
#we can also change the range parameter according to our needs
plt.hist(blood_sugar,bins=[80,100,150,200],rwidth=0.95)
```

```python
#Can also add labels over the bars
values, bins, bars=plt.hist(blood_sugar,bins=[80,100,150,200],rwidth=0.95)
plt.bar_label(bars,labels=['healthy','concerning','risky'])
```

```python
# Add two different data into the histogram
blood_sugar_men=[113,85,90,150,149,88,115,135,80,77,82,129]
blood_sugar_women=[67,98,89,120,133,150,84,69,89,79,120,112,100]
plt.xlabel('blood sugar range')
plt.ylabel('number of patients')
plt.title('blood sugar analysis')
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,color=['y','b'],label=['men','women'])
plt.legend()
```

```python
#to plot histogram in horizontal fashion we use orientation parameter
# Add two different data into the histogram
blood_sugar_men=[113,85,90,150,149,88,115,135,80,77,82,129]
blood_sugar_women=[67,98,89,120,133,150,84,69,89,79,120,112,100]
plt.ylabel('blood sugar range')
plt.xlabel('number of patients')
plt.title('blood sugar analysis')
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,color=['y','b'],label=['men','women'],orientation='horizontal')
plt.legend()
```

PIE CHART tells us the percentage of data that we are using

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals=[1400,600,300,410,250]
exp_labels=["Home","Food","PhoneBill","Car","Others"]
```

```python
plt.pie(exp_vals,labels=exp_labels)
```

```python
#to increase size of pie chart increase the size
plt.pie(exp_vals,labels=exp_labels,radius=1.5)#i.5 times the original size
plt.show()
```

```python
#to get the percentage we use autopct if we want digits after decimal write the places for it
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%0.2f%%')
plt.show()
```

```python
#to get one of the part to highlight(or to come out) it starts with first one
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%.2f%%',explode=[0,0.3,0,0,0])
plt.show()
```

```python
#to add home rent at to particular angle we add startangle in it
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%0.2f%%',explode=(0,0.2,0,0,0),startangle=180)
plt.show()
```

```python
#to save a chart in file format
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%0.2f%%',explode=(0,0.2,0,0,0),startangle=180)

plt.savefig('piechart.png',bbox_inches='tight',pad_inches=2)
```
