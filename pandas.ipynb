to remove na from file use function

```python
import pandas as pd
def convert_na(cell):
  if(cell=='n.a.'):
    return 'sam walton'
  return cell
def convert_not_available(cell):
  if(cell=='not available'):
    return 'None'
  return cell
def convert_negative(cell):
  if(cell==-1):
    return 0
  return cell

df=pd.read_excel('stock_data.xlsx','Sheet1',converters={
    'people':convert_na,
    'eps':convert_not_available,
    'price':convert_negative,
    'revenue':convert_negative
})
df

```

```python
#write back to data excel file
df.to_excel('new.xlsx',sheet_name='stocks')
```

If you have two data rame and you want to write two data frame to same excel file in different sheets then use ExcelWrite()

```python
import pandas as pd
df_stocks=pd.DataFrame({
    'tickers':['GOOGL','WMT','MSFT'],
    'price':[845,65,64],
    'pe':[30.37,14.26,30.97],
    'eps':[27082,4.61,2.12]
})
df_weather=pd.DataFrame({
    'day':['1/1/2017','1/2/2017','1/3/2017'],
    'temperature':[32,35,28],
    'windspeed':[6,7,2],
    'event':['Rain','Sunny','Snow']
})
with pd.ExcelWriter('stocks_weather.xlsx') as writer:
  df_stocks.to_excel(writer,sheet_name='stocks')
  df_weather.to_excel(writer,sheet_name='weather')
```

USING MERGE FOR DATFRAME

in case of concat we need to explicitly mention the index in merge it automatically detects and joins them in correct way


```python
import pandas as pd
df1=pd.DataFrame({
    "city":['new york','chicago','orlando','baltimore'],
    'temperature':[21,14,35,32]
})
df1
```

```python
df2=pd.DataFrame({
    "city":['chicago','new york','sanfranciso'],
    'humidity':[65,68,75]
})
df2
```

```python
df3=pd.merge(df1,df2)#similar to inner join in dbms
df3
```

image.png

```python
df4=pd.merge(df1,df2,on='city',how='outer')#outer join
df4
```

image.png

```python
#in order to get from which part of dataset we got the values we use indicator
df4=pd.merge(df1,df2,on='city',how='outer',indicator=True)
df4
```

```python
df5=pd.merge(df1,df2,on='city',how='left')#left join
df5
```

image.png

```python
df6=pd.merge(df1,df2,on='city',how='right')
df6
```

image.png

IF we have same column name in both the dataframe so to separate them we use suffixes


```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
df3= pd.merge(df1,df2,on="city",how="outer", suffixes=('_first','_second'))
df3
```
